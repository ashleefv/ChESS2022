##Introduction

Jupyter Notebook is a web-based tool that lets you create interactive notebooks using over 40 programming languages, including Python, R, Julia, and Scala. These interactive notebooks enable us to combine formatted text, computer code, interactive inputs and outputs, and other features, in a single publishable file. Live Scripts fall under a programming paradigm called [Literate Programming](https://https://en.wikipedia.org/wiki/Literate_programming) whereby a computer program or script is described/documented in a natural language (e.g., English) with formatted text and (interactive) code snippets. This tutorial will cover how to use [Google Colab](https://https://colab.research.google.com/) to create Jupyter Notebooks coded in Python.

###Code Cells

Code cells contain segments of executable code. On the top toolbar, the "Runtime" tab can be used to execute code cells either individually or as a group. Selecting the triangular execute button or entering `Ctrl + Enter` while selecting a code cell will also execute the code within that cell.

The sequence in which code cells execute is important because each code cell is only "aware" of what happened in previously executed code cells. Unlike MATLAB Live Scripts, a common function must be executed before it can be referenced in other code cells and a function must be confined to a single code cell to avoid execution errors.
